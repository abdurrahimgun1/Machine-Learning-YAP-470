```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
data = data.drop('price_range', axis=1)
correlation_matrix = data.corr()
correlation_matrix
```

```python
high_correlation_pairs = []
num_rows, num_cols = correlation_matrix.shape
for i in range(num_cols):
    for j in range(i+1, num_cols):
        correlation = correlation_matrix.iloc[i, j]
        if abs(correlation) > 0.9:
            pair = (correlation_matrix.columns[i], correlation_matrix.columns[j], correlation)
            high_correlation_pairs.append(pair)

for pair in high_correlation_pairs:
    print(f"Ã–znitelikler: {pair[0]}, {pair[1]} - Korelasyon: {pair[2]}")
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
correlation_matrix = data.corr()
correlation_matrix
```

```python
low_correlation_features = []
price_range_index = correlation_matrix.columns.get_loc('price_range')
for i in range(len(correlation_matrix)):
    correlation = correlation_matrix.iloc[i, price_range_index]
    if abs(correlation) < 0.1 and i != price_range_index:
        feature_name = correlation_matrix.columns[i]
        low_correlation_features.append((feature_name, correlation))

print("DÃ¼ÅŸÃ¼k Korelasyonlu Ã–znitelikler:")
for feature, correlation in low_correlation_features:
    print(f"{feature} - Korelasyon: {correlation}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x = data.drop(['price_range', 'blue', 'clock_speed', 'm_dep', 'n_cores', 'dual_sim', 'fc', 'four_g', 'int_memory', 'mobile_wt', 'pc', 'sc_h',
               'sc_w', 'talk_time', 'three_g', 'touch_screen', 'wifi'], axis=1)
y = data[["price_range"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(x_train, y_train)
```

```python
y_pred = rf_classifier.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Test doÄŸruluÄŸu:", accuracy)
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear')
```

```python
svm_model.fit(x_train, y_train)
```

```python
y_pred = svm_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("SVM modelinin doÄŸruluk skoru: {:.2f}".format(accuracy))
```
